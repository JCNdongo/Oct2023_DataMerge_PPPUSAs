### Merging all Contracting Data from 2018 to 2019 in one file ###

```python
#importing dependencies
import pandas as pd
```

```python
#define the directory path where each original file is located
directory_path = "F:\FAU PhD\DISSERTATION\DATABASES\DataTESTS\Oct.23.2023.Test\\"
```

```python
#read each original csv file into a a Pandas DataFrame
df_2018 = pd.read_csv(directory_path + "All_2018_Contracting.csv")
df_2019 = pd.read_csv(directory_path + "All_2019_Contracting.csv")
```

```python
#Merge the data from all the DataFrames based on the same "businessuei" data for every year. 
merged_df = df_2018.merge(df_2019, on='Businessuei', how='outer')
```

```python
# Remove duplicates based on 'businessuei'
merged_df.drop_duplicates(subset='Businessuei', keep='first', inplace=True)
merged_df.head()
```
